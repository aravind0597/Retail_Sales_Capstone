```python
import psycopg2

# Reconnect function
def get_connection():
    return psycopg2.connect(
        host="localhost",
        database="retail_sales_cap",
        user="postgres",
        password="1234"
    )

try:
    # Step 1: Establish the connection
    connection = get_connection()

    # Step 2: Create a cursor object
    mediator = connection.cursor()

    # Step 3: Add the `Order_id` column to `retail_sales2`
    mediator.execute('ALTER TABLE retail_sales2 ADD COLUMN IF NOT EXISTS "Order_id" BIGINT;')
    print("Added 'Order_id' column to retail_sales2.")

    # Step 4: Fetch `Order_id` values from `retail_sales1`
    mediator.execute('SELECT "Order_id" FROM retail_sales1;')
    retail_sales1_data = mediator.fetchall()

    # Step 5: Fetch all rows from `retail_sales2`
    mediator.execute("SELECT * FROM retail_sales2;")
    retail_sales2_data = mediator.fetchall()

    # Ensure row counts match (this is optional, depending on your requirement)
    if len(retail_sales1_data) != len(retail_sales2_data):
        raise ValueError("Row counts in retail_sales1 and retail_sales2 do not match!")

    # Step 6: Update the `Order_id` column in `retail_sales2` based on row order
    for idx, row in enumerate(retail_sales2_data):
        order_id = retail_sales1_data[idx][0]  # Get the corresponding `Order_id`
        update_query = 'UPDATE retail_sales2 SET "Order_id" = %s WHERE "Product_id" = %s;'  # Use Product_id for identification
        product_id = row[1]  # Assuming "Product_id" is the second column in retail_sales2
        mediator.execute(update_query, (order_id, product_id))  # Update Order_id based on Product_id

    # Commit the changes
    connection.commit()
    print("Database changes committed successfully.")

except psycopg2.Error as e:
    print(f"Database Error: {e}")
    connection.rollback()

except ValueError as ve:
    print(f"Error: {ve}")
    connection.rollback()

finally:
    # Close the cursor and connection
    if 'connection' in locals() and not connection.closed:
        mediator.close()
        connection.close()
        print("Connection closed properly.")
```

```python
import psycopg2

connection = psycopg2.connect(
    host = "localhost",
    user = "postgres",
    password = "1234",
    database = "retail_sales_cap"
)

mediator = connection.cursor()

mediator.execute('ALTER table retail_sales1 add constraint Order_id_pk PRIMARY KEY ("Order_id")')
print("Primary Key added successfully")

mediator.execute('ALTER table retail_sales2 add constraint Order_id_fk FOREIGN KEY ("Order_id") REFERENCES retail_sales1 ("Order_id")')
print("Foreign key added successfully")

connection.commit()
print("Database changes committed successfully.")

mediator.close()
connection.close()
print("Connection closed properly.")

```
